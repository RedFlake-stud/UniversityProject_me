```python
import tkinter as tk
from tkinter import *

# Medical staff ir Room gal ir hospital klases mesti i atskirus failus

class Medical_staff:
    def __init__(self, name, age, gender):
        self._name = name
        self._age = age
        self._gender = gender

class Doctor(Medical_staff):
    def __init__(self, name, age, gender, specialization):
        super().__init__(name,age,gender)
        self._specialization = specialization
        self._status = "Available"

    def update_status(self, new_status):
        self._status = new_status

class Nurse(Medical_staff):
    def __init__(self, name, age, gender, specialization):
        super().__init__(name,age,gender)
        self._status = "Available"

    def update_status(self, new_status):
        self._status = new_status


class Room:
    def __init__(self, room_number, capacity, patients):
        self._room_number = room_number
        self._capacity = capacity
        self._is_occupied = False
        self._patient_amount = patients

    def occupy_room(self, patient):
        if self._patient_amount >= self.capacity:
            print("Room is occupied.")
        else:
            print("Room has space left.")
        print(f"{self._patient_amount}/{self._capacity}")
        self.update_is_occupied()

    def update_patient_amount(self, new_patient_amount):
        self._patient_amount = new_patient_amount

    def update_is_occupied(self):
        if self._patient_amount >= self.capacity:
            self._is_occupied = True
        else:
            self._is_occupied = False


class Hospital:
    def __init__(self, name):
        self._name = name
        self._doctors = []
        self._nurses = []
        self._rooms = []

    def add_doctor(self, doctor):
        for i in self._doctors:
            if i._name == doctor._name:
                return "Doctor already exists"
            else:
                self._doctors.append(doctor)

    def add_nurse(self, nurse):
        for i in self._nurses:
            if i._name == nurse._name:
                return "Nurse already exists"
            else:
                self._nurses.append(nurse)

    def add_room(self, room):
        self._rooms.append(room)

    def remove_doctor(self, doctor):
        self._doctors.remove(doctor)

    def remove_nurse(self, nurse):
        self._nurses.remove(nurse)

    def remove_room(self, room):
        self._rooms.remove(room)

    def decorator(self, func):
        if func == "add_doctor":
            return self.add_doctor
        elif func == "add_nurse":
            return self.add_nurse
        elif func == "add_room":
            return self.add_room
        elif func == "remove_doctor":
            return self.remove_doctor
        elif func == "remove_nurse":
            return self.remove_nurse
        elif func == "remove_room":
            return self.remove_room
        else:
            return None


hospital = Hospital("Hospital")

m = tk.Tk()

m.geometry("750x500")
m.title("Emergency Helper")

label = tk.Label(m, text = "Start", font = ('Arial', 30))
label.pack(pady = 100)


buttonframe = tk.Frame(m)
buttonframe.columnconfigure(0, weight = 1)
buttonframe.columnconfigure(1, weight = 1)

btn1 = tk.Button(buttonframe, text = "Emergency Mode", font = ('Arial', 20))
btn1.grid(row = 0, column = 0, sticky = tk.W + tk.E)

btn2 = tk.Button(buttonframe, text = "Configure Hospital", font = ('Arial', 20))
btn2.grid(row = 0, column = 1, sticky = tk.W + tk.E)

buttonframe.pack(fill = 'x')

anotherbtn = tk.Button(m, text = "Exit")
anotherbtn.place(x = 325, y = 350, height = 100, width = 100)

m.mainloop()
```
